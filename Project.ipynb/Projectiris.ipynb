# Summary of each variable
The write function will create the text file named 'iris_summary_stats.txt' where the summary of each variable will be saved

```python
sepal_len = data['sepal_length'].describe()
print(sepal_len)

#sepal_len.to_csv('iris_summary_stats.txt', sep='\t')
with open('iris_summary_stats.txt', 'w') as fl:
    fl.write(sepal_len.to_csv(sep='\t')+'\n')
```

```python
sepal_wid = data['sepal_width'].describe()
print(sepal_wid)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(sepal_wid.to_csv(sep='\t')+'\n')
```

```python
petal_len = data['petal_length'].describe()
print(petal_len)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(petal_len.to_csv(sep='\t')+'\n')
```

```python
petal_wid = data['petal_width'].describe()
print(petal_wid)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(petal_wid.to_csv(sep='\t')+'\n')
```
